```python
def dSdy_Plots(Si):
    plot_dict = {}
    for fld in fds.y.fields:
        plot_dict[fld] = []
        for smb in fld.s:
            if smb in fds.y.dynamic_s:
                plot_dict[fld].append(Si.dSdy[smb].xreplace(fds.y.args))
            else:
                plot_dict[fld].append(0)
                
    # Create individual plots and capture them in Output widgets
    metric = widgets.Output()
    with metric:
        fig1, ax1 = plt.subplots(figsize=(4, 3))
        ax1.plot(fds.f.x_list, plot_dict[fds.f], color = "red", label = f"$dS/df$ value")
        ax1.plot(fds.h.x_list, plot_dict[fds.h], color = "blue", label = f"$dS/dh$ value")

        ax1.legend(loc = "upper right", fontsize="small")
        ax1.set_title(f"Ansatz vs. Solution: $f(x)$ & $h(x)$");
        ax1.set_xlabel("Rescaled Position $x$");
        
        display(fig1)
        plt.close(fig1) # Close the figure to prevent immediate display
    
    conformalon = widgets.Output()
    with conformalon:
        fig2, ax2 = plt.subplots(figsize=(4, 3))
        ax2.plot(fds.ph.x_list, plot_dict[fds.ph], color = "red", label = fr"$dS/d \phi$ value")
        ax2.plot(fds.psi.x_list, plot_dict[fds.psi], color = "blue", label = fr"$dS/d \psi$ value")
        
        ax2.legend(loc = "upper right", fontsize="small")
        ax2.set_title(f"Ansatz vs. Solution for $\\phi(x)$ & $\\psi(x)$");
        ax2.set_xlabel("Rescaled Position $x$");

        display(fig2)
        plt.close(fig2) # Close the figure to prevent immediate display
    
    fourform = widgets.Output()
    with fourform:
        fig3, ax3 = plt.subplots(figsize=(4, 3))
        ax3.plot(fds.f.x_list, plot_dict[fds.w], color = "red", label = f"$dS/dw$ value")
        
        #plt.legend(loc = "upper right")
        ax3.set_title(f"Ansatz vs. Solution for $w(x)$");
        ax3.set_xlabel("Rescaled Position $x$");
        
        ax3.legend(loc = "lower right", fontsize="small")
        display(fig3)
        plt.close(fig3)
    
    # Display them horizontally using HBox
    box = widgets.HBox([metric, conformalon, fourform], layout=widgets.Layout(justify_content="space-around"))
    
    display(box)
```

```python
def twiddle_point(smb, step, expr_list, avoid_dd = True):
    return_string = "stuck"
    idx, fld = fds.y.idx_dict[smb]
    starting_sdd = abs(fld.sdd[idx].xreplace(fds.y.args))
    if avoid_dd:
        if fld == fds.f:
            max_sdd = .25
        elif fld == fds.b:
            max_sdd = .0025
        else:
            max_sdd = .0025
            print("What is the field???")
    else:
        max_sdd = 100
    BrandNew = True
    old_val = fld.v[idx]

    Sq = []
    Start_String = "Start:"
    for expr, label in expr_list:
        Sq_val = expr.xreplace(fds.y.args)
        Sq.append(Sq_val)
        Start_String += f" {label}={Sq_val:.4f} "
    Start_String += f"{fld.label}[{idx}] = {fds.y.args[fld.s[idx]]:.4f}"
    print(Start_String)
    
    while True:
        fld.v[idx] = old_val + step        
        fds.y.update_args()
        S.runNR(50, dynamic = dynamic_list, lambdify_first = True, PureNR = True)
        Sq_new = []
        Update_String = "Update to:"
        for expr, label in expr_list:
            Sq_val = expr.xreplace(fds.y.args)
            Sq_new.append(Sq_val)
            Update_String += f" {label}={Sq_val:.4f} "
        Update_String += f"{fld.label}[{idx}] = {fds.y.args[fld.s[idx]]:.4f}"
        print(Update_String)
        if all(new <= old for new, old in zip(Sq_new, Sq)):
            print("Accepted")
            return_string = "moved"
            BrandNew = False
            Sq = Sq_new.copy()
            old_val = fld.v[idx]
            if abs(fld.sdd[idx].xreplace(fds.y.args)) >= max_sdd and abs(fld.sdd[idx].xreplace(fds.y.args)) >= starting_sdd:
                print("Stopped On Account of Too High Second Derivative")
                return step, "moved"
        elif BrandNew and all(new >= old for new, old in zip(Sq_new, Sq)):
            print("Rejected.  Maybe switch directions?")
            BrandNew = False
            step = -1*step
            fld.v[idx] = old_val
            fds.y.update_args()
        else:
            print("Fully Rejected")
            step = step * -1
            fld.v[idx] = old_val
            fds.y.update_args()
            S.runNR(50, dynamic = dynamic_list, lambdify_first = True, PureNR = True)
            
            Sq = []
            Final_String = "Final:"
            for expr, label in expr_list:
                Sq_val = expr.xreplace(fds.y.args)
                Sq.append(Sq_val)
                Final_String += f" {label}={Sq_val:.4f} "
            Final_String += f"{fld.label}[{idx}] = {fds.y.args[fld.s[idx]]:.4f}"
            print(Final_String)

            return step, return_string

def add_smb_to_dynamic(smb):
    if smb not in dynamic_list:
        dynamic_list.append(smb)
        worked = S.runNR(30, dynamic = dynamic_list, lambdify_first = True, PureNR = False)
    else:
        print("Already in List")
        return dynamic_list
    if worked:
        return dynamic_list, worked
    else:
        worked = False
        dynamic_list.remove(smb)
        return dynamic_list, worked  
```
