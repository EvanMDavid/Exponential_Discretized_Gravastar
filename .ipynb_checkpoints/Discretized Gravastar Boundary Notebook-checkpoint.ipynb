```python
def twiddle_point(smb, step, expr_list, avoid_dd = True):
    return_string = "stuck"
    idx, fld = fds.y.idx_dict[smb]
    starting_sdd = abs(fld.sdd[idx].xreplace(fds.y.args))
    if avoid_dd:
        if fld == fds.f:
            max_sdd = .25
        elif fld == fds.b:
            max_sdd = .0025
        else:
            max_sdd = .0025
            print("What is the field???")
    else:
        max_sdd = 100
    BrandNew = True
    old_val = fld.v[idx]

    Sq = []
    Start_String = "Start:"
    for expr, label in expr_list:
        Sq_val = expr.xreplace(fds.y.args)
        Sq.append(Sq_val)
        Start_String += f" {label}={Sq_val:.4f} "
    Start_String += f"{fld.label}[{idx}] = {fds.y.args[fld.s[idx]]:.4f}"
    print(Start_String)
    
    while True:
        fld.v[idx] = old_val + step        
        fds.y.update_args()
        S.runNR(50, dynamic = dynamic_list, lambdify_first = True, PureNR = True)
        Sq_new = []
        Update_String = "Update to:"
        for expr, label in expr_list:
            Sq_val = expr.xreplace(fds.y.args)
            Sq_new.append(Sq_val)
            Update_String += f" {label}={Sq_val:.4f} "
        Update_String += f"{fld.label}[{idx}] = {fds.y.args[fld.s[idx]]:.4f}"
        print(Update_String)
        if all(new <= old for new, old in zip(Sq_new, Sq)):
            print("Accepted")
            return_string = "moved"
            BrandNew = False
            Sq = Sq_new.copy()
            old_val = fld.v[idx]
            if abs(fld.sdd[idx].xreplace(fds.y.args)) >= max_sdd and abs(fld.sdd[idx].xreplace(fds.y.args)) >= starting_sdd:
                print("Stopped On Account of Too High Second Derivative")
                return step, "moved"
        elif BrandNew and all(new >= old for new, old in zip(Sq_new, Sq)):
            print("Rejected.  Maybe switch directions?")
            BrandNew = False
            step = -1*step
            fld.v[idx] = old_val
            fds.y.update_args()
        else:
            print("Fully Rejected")
            step = step * -1
            fld.v[idx] = old_val
            fds.y.update_args()
            S.runNR(50, dynamic = dynamic_list, lambdify_first = True, PureNR = True)
            
            Sq = []
            Final_String = "Final:"
            for expr, label in expr_list:
                Sq_val = expr.xreplace(fds.y.args)
                Sq.append(Sq_val)
                Final_String += f" {label}={Sq_val:.4f} "
            Final_String += f"{fld.label}[{idx}] = {fds.y.args[fld.s[idx]]:.4f}"
            print(Final_String)

            return step, return_string

def add_smb_to_dynamic(smb):
    if smb not in dynamic_list:
        dynamic_list.append(smb)
        worked = S.runNR(30, dynamic = dynamic_list, lambdify_first = True, PureNR = False)
    else:
        print("Already in List")
        return dynamic_list
    if worked:
        return dynamic_list, worked
    else:
        worked = False
        dynamic_list.remove(smb)
        return dynamic_list, worked  
```

```python
ln = smp.log

x, xm, xp = smp.symbols("x x_- x_+")

alpha, beta, gamma = smp.symbols("alpha, beta, gamma")
eps = smp.symbols("epsilon")
cS, cinf, c0, cH = smp.symbols("c_S, c_\infty, c_0, c_H")
cw1, cw2 = smp.symbols("c_w1, c_w2")

aF = smp.Function("a_F")(x)
u = 1 - smp.sqrt(x**2/10000)
aF_val = 1/8*(2*u*(u**2 + 3) + 3*(u**4 - 1)*ln((1 + u)/(1 - u))) + .053

f = smp.Function("f")(x)
b = smp.Function("b")(x)
w = smp.Function("w")(x)
phi = smp.Function("varphi")(x)
psi = smp.Function("psi")(x)

fp = smp.diff(f, x)
bp = smp.diff(b, x)
wp = smp.diff(w, x)
wpp = smp.diff(w, (x, 2))

phip = smp.diff(phi, x)
psip = smp.diff(psi, x)

vals_dict = {gamma : 1, aF : aF_val, cH : 1, cS : 1, alpha : 1, c0 : 1, cinf : 1 + 56.6, alpha : 1, eps : 1e-4, cw1 : 0, cw2 : 0}


LEH = 1/b - 2 * f * bp - b * fp
LF = -3/4 * (b * wp**2) - 3/2*(w * bp * wp) - 3/2 * w * b * wpp
Lint = -aF * gamma * w * (phip + psip)

box2 = lambda fld: b**2 * f * smp.diff(fld, (x, 2)) + smp.diff(fld, x) * (b*bp*f + b**2*fp)
R = - b * smp.diff(b*fp, x)

L4 = alpha * 1/b * (box2(phi)**2 - b**2*f*R/3 * phip**2 + 2/3*R*box2(phi))
L4 = L4.simplify()
L5 = beta * 1/b * R**2 * phi
L6 = -1 * alpha * 1/b * (box2(psi)**2 - b**2*f*R/3 * psip**2)
L6 = L6.simplify()

Ltot = LEH + LF + Lint + L4 + L5 + L6

def CreateEoM(L, fld):
    EoM = 0
    k = 0
    sign = 1
    while k < 4:
        wrt = smp.diff(fld, (x, k))
        EoM += sign * smp.diff(smp.diff(L, wrt), (x, k))
        sign = sign * -1
        k += 1
    return EoM

def CalcEoM(x, EoM, dictionary):
    functions = list(EoM.atoms(AppliedUndef))
    derivatives = list(EoM.atoms(smp.Derivative))
    deriv_order = 4
    while deriv_order > 0:
        for deriv in derivatives:
            if deriv.derivative_count == deriv_order:
                deriv_func = deriv.expr
                if deriv_func in dictionary.keys():
                    deriv_val = dictionary[deriv_func].derivative(deriv_order)(x)
                    EoM = EoM.xreplace({deriv : deriv_val})
                else:
                    print(f"{deriv_func} not in keys")
        deriv_order -= 1        
    for func in functions:
        if func in dictionary.keys():
            func_val = dictionary[func](x)
            EoM = EoM.xreplace({func : func_val})
    return EoM
```

```python
def w_var(i):
    return_val = smp.sqrt(h.s[i]/f.s[i])*w.sd[i] + (2*gamma/3)*(ph.sd[i] + 2*psi.sd[i])
    return return_val

def phi_var(i):
    return_val = dAlembert(dAphi, i) + (1/3)*R.s[i]*dAphi.s[i] + (1/3)*h.s[i]*R.sd[i]*ph.sd[i] + (1/3)*dAlembert(R, i) + (beta/(2*alpha))*(R.s[i])**2 + gamma/(2*alpha)*smp.sqrt(h.s[i]/f.s[i])*w.sd[i]
    return return_val.xreplace({Dx : dx})

def psi_var(i):
    return_val = dAlembert(dApsi, i) + (1/3)*R.s[i]*dApsi.s[i] + (1/3)*h.s[i]*R.sd[i]*psi.sd[i] - gamma/(2*alpha)*smp.sqrt(h.s[i]/f.s[i])*w.sd[i]
    return return_val

def rho1(field, i):
    if field == ph:
        dA = dAphi
    elif field == psi:
        dA = dApsi
    else:
        print("rho function had trouble with field")
    return (2/3)*(R.s[i]*dA.s[i] + h.s[i]*field.sd[i]*R.sd[i] + 2*h.s[i]*dA.sdd[i] + h.sd[i]*dA.sd[i])

def rho2(field, i):
    if field == ph:
        dA = dAphi
    elif field == psi:
        dA = dApsi
    else:
        print("rho function had trouble with field")
    if i + 1 == len(field.sdd):
        dddf = (field.sdd[i] - field.sdd[i - 1])/Dx
    else:
        dddf = (field.sdd[i+1] - field.sdd[i])/Dx
    rho2a = dAlembert(field, i)**2 + 2*h.s[i]*field.sd[i]*dA.sd[i]
    rho2b = - (1/3)*(2*h.s[i]*(h.sdd[i]*field.sd[i]**2 + 4*h.sd[i]*field.sd[i]*field.sdd[i] + 2*h.s[i]*(field.sdd[i])**2 + 2*h.s[i]*field.sd[i]*dddf) + h.sd[i]*(h.sd[i]*field.sd[i]**2 + 2*h.s[i]*field.sd[i]*field.sdd[i]))
    return rho2a + rho2b

def rhoC(field, i):
    return R.s[i]**2*field.s[i] + 2*(2*h.s[i]*(R.sdd[i]*field.s[i] + 2*R.sd[i]*field.sd[i] + R.s[i]*field.sdd[i]) + h.sd[i]*(R.sd[i]*field.s[i] + R.s[i]*field.sd[i]))

def P1(field, i):
    if field == ph:
        dA = dAphi
    elif field == psi:
        dA = dApsi
    else:
        print("P function had trouble with field")
    P1 = (2/3)*(-1*R.s[i]*dA.s[i] + h.s[i]*field.sd[i]*R.sd[i] - (h.s[i]/f.s[i])*f.sd[i]*dA.sd[i])
    return P1

def P2(field, i):
    if field == ph:
        dA = dAphi
    elif field == psi:
        dA = dApsi
    else:
        print("P function had trouble with field")
    P2 = -(dA.s[i])**2 + 2*h.s[i]*field.sd[i]*(dA.sd[i]) + (1/3)*(2*R.s[i]*h.sd[i]*(field.sd[i])**2 + (h.s[i]/f.s[i])*f.sd[i]*(h.sd[i]*(field.sd[i])**2 + 2*h.s[i]*field.sd[i]*field.sdd[i]))
    return P2

def PC(field, i):
    return 0

def f_var(i):
    return 1 - h.sd[i] - (3/4)*(h.s[i]/f.s[i])*(w.sd[i])**2 - alpha*(rho1(ph, i) + rho2(ph, i) - 2*rho2(psi, i)) - beta*rhoC(ph, i)

def h_var(i):
    return (h.s[i]/f.s[i])*f.sd[i] - 1 + (3/4)*(smp.sqrt(h.s[i]/f.s[i])*w.sd[i]) - alpha*(P1(ph, i) - P2(psi, i)) - beta*PC(ph, i)
```

```python
idx = 7
n = idx

w = fds.w.s
f = fds.f.s
h = fds.h.s
phi = fds.ph.s
psi = fds.psi.s

v = [smp.sqrt(f[n]/h[n]) for n in range(len(f))]
dAphi = [fds.dAlembert(fds.ph, n) for n in range(len(f))]
dApsi = [fds.dAlembert(fds.psi, n) for n in range(len(f))]
R = [fds.R_func(n) for n in range(len(f))]

#Emils Expressions

def L4(n):
    expr = alpha*v[n]*( dAphi[n]**2 - 1/3*h[n]*R[n]*((phi[n+1] - phi[n])/Dx)**2 + 2/3*R[n]*dAphi[n] )
    return expr

def L6(n):
    expr = -2*alpha*v[n]*( dApsi[n]**2 - h[n]*R[n]/3*((psi[n+1] - psi[n])/Dx)**2)
    return expr

def dSdw(n):
    expr1 = 3/(2*Dx**2)*((w[n] - w[n-1])/v[n - 1] - (w[n+1] - w[n])/v[n])
    expr2 = gamma/Dx * (phi[n-1] + 2 * psi[n-1] - phi[n] - 2*psi[n])
    return smp.expand(Dx*(expr1 + expr2))

def dSdph(n):
    expr1 = gamma*(w[n+1] - w[n])/Dx
    expr2 = alpha*(v[n-1]/Dx**2)*((dAphi[n-1] + 1/3*R[n-1])*(h[n] + h[n-1]*f[n]/f[n-1]) - 2/3*h[n-1]*R[n-1]*(phi[n] - phi[n-1]))
    expr3 = -alpha*(v[n]/Dx**2)*((dAphi[n] + (1/3)*R[n])*(h[n+1] + 2*h[n] + h[n]*f[n+1]/f[n]) - 2/3*h[n]*R[n]*(phi[n+1] - phi[n]))
    expr4 = 2*alpha*(v[n+1]/Dx**2)*(dAphi[n+1] + 1/3*R[n+1])*h[n+1]
    expr5 = beta*smp.sqrt(f[n]/h[n])*(R[n])**2

    return smp.expand((expr1 + expr2 + expr3 + expr4 + expr5)*Dx)

def dSdpsi(n):
    expr1 = -2*alpha*(v[n-1]/Dx**2)*(dApsi[n-1]*(h[n] + h[n-1]*f[n]/f[n-1]) - 2/3*h[n-1]*R[n-1]*(psi[n] - psi[n-1]))
    expr2 = 2*alpha * (v[n]/Dx**2) * ( dApsi[n]*(h[n+1] + 2*h[n] + h[n]*f[n+1]/f[n]) - 2/3*h[n]*R[n]*(psi[n+1] - psi[n]) )
    expr3 = -4*alpha*v[n+1]/Dx**2*dApsi[n+1]*h[n+1]
    expr4 = 2*gamma*(w[n+1] - w[n])/Dx

    return Dx*(expr1 + expr2 + expr3 + expr4)

def dSdf(n):
    dL1nm = 0
    dL1n = 1/(2*smp.sqrt(f[n]*h[n]))*(1 - (h[n+1] - h[n])/Dx)
    dL1np = 0
    dL2nm = 0
    dL2n = -3/8*smp.sqrt(h[n]/f[n]**3)*((w[n+1] - w[n])/Dx)**2
    dL2np = 0
    dL4nm = alpha*v[n-1]*( 2*(dAphi[n-1] + 1/3*R[n-1])*smp.diff(dAphi[n-1], f[n]) + 1/3*(2*dAphi[n-1] - h[n-1]*((phi[n] - phi[n-1])/Dx)**2)*smp.diff(R[n-1], f[n]))
    dL4n = smp.diff(v[n], f[n])*L4(n)/v[n] + alpha*v[n]*( 2*(dAphi[n] + 1/3*R[n])*smp.diff(dAphi[n], f[n]) + 1/3*(2*dAphi[n] - h[n]*((phi[n+1] - phi[n])/Dx)**2)*smp.diff(R[n], f[n]))
    dL4np = -alpha*v[n+1]/(3*Dx**2)*(2*dAphi[n+1] - h[n+1]*((phi[n+2] - phi[n+1])/Dx)**2)*(h[n+1]/f[n+1])
    dL5nm = 2*beta*v[n-1]*R[n-1]*phi[n-1]*smp.diff(R[n-1], f[n])
    dL5n = beta*v[n]*R[n]*phi[n]*( R[n]/(2*f[n]) + 2*smp.diff(R[n], f[n]) )
    dL5np = -2*beta/Dx**2*smp.sqrt(h[n+1]/f[n+1])*phi[n+1]*R[n+1]
    dL6nm = -2*alpha*v[n-1]*( 2*dApsi[n-1]*smp.diff(dApsi[n-1], f[n]) - 1/3*h[n-1]*((psi[n] - psi[n-1])/Dx)**2*smp.diff(R[n-1], f[n]) )
    dL6n = 1/(2*f[n])*L6(n) - 2*alpha*v[n]*( 2*dApsi[n]*smp.diff(dApsi[n], f[n]) - 1/3*h[n]*((psi[n+1] - psi[n])/Dx)**2*smp.diff(R[n], f[n]) )
    dL6np = -2*alpha*v[n+1]/3*(psi[n+2] - psi[n+1])**2/(Dx**4)*(h[n+1]**2/f[n+1])

    return smp.expand(Dx*(dL1n + dL2n + dL4nm + dL4n + dL4np + dL5nm + dL5n + dL5np + dL6nm + dL6n + dL6np))

def dSdh(n):
    dL1nm = -1/Dx*v[n-1]
    dL1n = v[n]/(2*h[n])*( (h[n+1] + h[n])/Dx - 1 )
    dL2n = 3/8*v[n]/f[n]*( ((w[n+1] - w[n])/Dx)**2 )
    dL4nm = alpha*v[n-1]*( 2*(dAphi[n-1] + 1/3*R[n-1])*smp.diff(dAphi[n-1], h[n]) + 1/3*(2*dAphi[n-1] - h[n-1]*((phi[n] - phi[n-1])/Dx)**2)*smp.diff(R[n-1], h[n]) )
    dL4n = alpha*v[n]*( 2*(dAphi[n] + 1/3*R[n])*smp.diff(dAphi[n], h[n]) + 1/3*(2*dAphi[n] - h[n]*((phi[n + 1] - phi[n])/Dx)**2)*smp.diff(R[n], h[n]) - 1/(2*h[n])*dAphi[n]*(dAphi[n] + 2/3*R[n]) - 1/6*R[n]*((phi[n+1] - phi[n])/Dx)**2 )
    dL5nm = 2*beta*v[n-1]*R[n-1]*phi[n-1]*smp.diff(R[n-1], h[n])
    dL5n = beta*v[n]*R[n]*phi[n]*(2*smp.diff(R[n], h[n]) - 1/(2*h[n])*R[n])
    dL6nm = -2*alpha*v[n-1]*( 2*dApsi[n-1]*smp.diff(dApsi[n-1], h[n]) - 1/3*h[n-1]*((psi[n] - psi[n-1])/Dx)**2*smp.diff(R[n-1], h[n]) )
    dL6n = -2*alpha*v[n]*( 2*dApsi[n]*smp.diff(dApsi[n], h[n]) - 1/3*h[n]*((psi[n + 1] - psi[n])/Dx)**2*smp.diff(R[n], h[n]) - 1/2*dApsi[n]**2/h[n] - 1/6*R[n]*((psi[n+1]-psi[n])/Dx)**2 )

    return smp.expand(Dx*(dL1nm + dL1n + dL2n + dL4nm + dL4n + dL5nm + dL5n + dL5np + dL6nm + dL6n))
```
