```python
def add_new_points(action, new_points = 1, double = True):
    global N
    N = action.fields.N
    global S
    if double:
        N += new_points*(N-3)
    else:
        N += new_points
    ansatz_list = []
    for fld in action.field_list:
        if fld.dynamic:
            ansatz_list.append(interp1d(fld.x_list, fld.v, kind = "cubic"))
        else:
            ansatz_list.append(fld.ansatz)

    fds.create_fields(ansatz_list, [-100, 100, N])
    S = Action(fds.Ltot, fds.y, lambdify_type = action.lambdify_type, params = action.params, lambdify_deriv = False)
```

```python
plot_all_fields()

S.popEM()
dSdy_old = S.E.copy()
eig = np.linalg.eig(S.M)
eig_val = eig.eigenvalues
eig_vec = eig.eigenvectors.T
dSdy_in_eig_basis = []
for e in eig_vec:
    dSdy_in_eig_basis.append(sum(e*S.E))
delta_in_eig_basis = -1*(dSdy_in_eig_basis/eig_val)

max_idx = list(delta_in_eig_basis).index(max(delta_in_eig_basis))
min_idx = list(delta_in_eig_basis).index(min(delta_in_eig_basis))

if abs(delta_in_eig_basis[max_idx]) > abs(delta_in_eig_basis[min_idx]):
    idx = max_idx
else:
    idx = min_idx

def dSdy_in_eig_basis_func(variables, eigenvector):
    return_val = 0
    k = 0
    for smb in fds.y.dynamic_s:
        return_val += eigenvector[k]*S.lambda_dSdy[smb](*variables)
        k+=1
    return return_val


step_in_biggest_direction = eig_vec[idx]*delta_in_eig_basis[idx]

S.runNR(1, dynamic = dynamic_list_l, PureNR = False, lambdify_first = False, revert_if_failed = False, output = False)
S.popEM()
dSdy_new = S.E.copy()

plot_all_fields()

plt.plot(fds.l.x_list[2:N-2], dSdy_old, label = "Old")
plt.plot(fds.l.x_list[2:N-2], dSdy_new, label = "New")
plt.xlabel("$dS/d\lambda$")
plt.legend()
plt.show()

plt.title("Newton Raphson Update")
plt.plot(fds.l.x_list[2:N-2], S.delta)
plt.show()

print(idx)
print(f"Eigenvalue of Largest Step {eig_val[idx]:.2f}")
print(f"dSdy of Largest Step {dSdy_in_eig_basis[idx]:.2f}")
```
